```
library(SportsAnalytics)
library(ggplot2)
```

```
NBA1920 <- fetch_NBAPlayerStatistics("19-20")
NBA1819 <- fetch_NBAPlayerStatistics("18-19")
NBA1718 <- fetch_NBAPlayerStatistics("17-18")
```

```
GSWdata1920 <- NBA1920[grepl("GSW", NBA1920$Team),]
GSWdata1819 <- NBA1819[grepl("GSW", NBA1819$Team),]
GSWdata1718 <- NBA1718[grepl("GSW", NBA1718$Team),]
```

盒型圖：由圖一可看出整個聯盟各位置的上場時間平均大致相等，而GSW的各位置上場時間差異相當顯著



```
ggplot(NBA1920, aes(x = Position, y = TotalMinutesPlayed)) + geom_boxplot()
```

```
ggplot(GSWdata1920, aes(x = Position, y = TotalMinutesPlayed)) + geom_boxplot()
```

散佈圖：
由圖之呈現可看出上場時間與總得分有相關性。且根據趨勢線可看出GSW的趨勢與聯盟趨勢相近

```
ggplot(NBA1920, aes(x = TotalMinutesPlayed, y = FieldGoalsMade)) + geom_point() + geom_smooth()
```

```
ggplot(GSWdata1920, aes(x = TotalMinutesPlayed, y = FieldGoalsMade)) + geom_point() + geom_smooth()
```
