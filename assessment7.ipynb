```
data <- read.csv("c45339a14f50eadb30c3b28140017011_export.csv")
```

```
data1 <- data[grepl("亞洲", data$`洲別`),]
```

```
ggplot(data1, aes(x = 學位生_正式修讀學位外國生, y = 非學位生_外國交換生)) + geom_point()
```

外國學生來台就讀並不會影響到外國學生來臺灣交換的意願

```
ggplot() + geom_bar(data = data1, aes(x = 國別, y = 學位生_正式修讀學位外國生), stat = "identity")
```

因某些國家的人數太多了，故從圖只能看見人數較多的幾個國家。

```
ggplot(data, aes(x = 洲別, y = 學位生_正式修讀學位外國生)) + geom_boxplot()
```
