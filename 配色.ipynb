```
library(SportsAnalytics)
library(ggplot2)
```

```
NBA1920 <- fetch_NBAPlayerStatistics("19-20")
```

1.利用order()函數,由大到小排序FieldGoalsMade

```
NBA1920Order <- NBA1920[order(NBA1920$FieldGoalsMade, decreasing = T),]
NBA1920Order
```

2.用subset()函數搜尋隊伍為"LAL"

```
subset(NBA1920,Team=="LAL")
```

3.用grepl()搜尋名字有"James"的球員

```
NBA1920[grepl("James",NBA1920$Name),]
```

4.用ggplot畫出,四種不同類型的圖表

(1)x為Position，y為FieldGoalsMade之盒狀圖。

```
ggplot(NBA1920, aes(x = Position, y = FieldGoalsMade)) + geom_point()
```

```
ggplot(NBA1920, aes(x = Position, y = FieldGoalsMade)) + geom_boxplot()
```

(2)x為TotalMinutesPlayedy，y為FieldGoalsMade之散布圖。可看出兩者有正相關。

```
ggplot(NBA1920, aes(x = TotalMinutesPlayed, y = FieldGoalsMade)) + geom_point() + geom_smooth()
```

(3)x為LAL之各球員姓名，y為其FieldGoalsMade之長條圖。

```
NBA1920LAL <- subset(NBA1920,Team=="LAL")
```

```
ggplot() + geom_bar(data = NBA1920LAL, aes(x = Name, y = FieldGoalsMade), stat = "identity")
```

(4)x為SG之GamesPlayed，y為其FieldGoalsMade之折線圖。可看出相同位置的FieldGoalsMade跟GamesPlayed無直接關係。

```
NBA1920SG <- subset(NBA1920,Position=="SG")
```

```
ggplot(NBA1920SG, aes(x = GamesPlayed, y = FieldGoalsMade)) + geom_line()
```
